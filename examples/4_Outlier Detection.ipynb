#Choose and sample outliers with interactive functions.

The following functions are introduced:

  * `study.interactive_choose_outliers`<br>
    This function uses one-class SVM to identify outliers based on user-chosen fitting parameters.
  <br>
  <br>
     
  * `study.interactive_reset_outliers`<br>
    User can combine multiple outlier calls to set `study.metadata.data['outlier']`
  <br>
  <br>

  * `study.drop_outliers`<br>
    After setting the `outlier` attribute in `study.metadata.data` the samples labelled outliers are removed from `study.expression.data` and `study.splicing.data`
